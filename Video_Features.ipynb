```python
from mmaction.apis import inference_recognizer, init_recognizer
import os
import numpy as np

config_path = 'configs/recognition/timesformer/timesformer_spaceOnly_8xb8-8x32x1-15e_kinetics400-rgb.py'
checkpoint_path = 'https://download.openmmlab.com/mmaction/recognition/timesformer/vit_base_patch16_224.pth' # can be a local path

#for loop for the videos
#Important: create this folder inside of the mmaction2 root directory
audio_folder = 'sample'
all_audio = [f for f in os.listdir(audio_folder) if f.endswith('.mp4')]
all_audio.sort()

for file in all_audio:
  print(type(file))
  model = init_recognizer(config_path, checkpoint_path, device="cpu")
  result = inference_recognizer(model, 'sample/'+str(file))
  #change directory to save
  save_file_name =  os.path.join('','./','{}_deep_video.'.format(file))
  np.save(save_file_name , np.array(result,dtype=object), allow_pickle=True)
  # print(result)
```
